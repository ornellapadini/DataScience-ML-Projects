```python
def segundos_vividos(fecha):
    '''
    Recibe una fecha de nacimiento como cadena en formato 'dd/mm/aaaa'
    (día, mes, año con 2, 2 y 4 dígitos, separados con barras normales)
    y devuelve la cantidad de segundos que viviste
    '''
    from datetime import datetime, timedelta
    date_object = datetime.strptime(fecha, '%d/%m/%Y')
    t0 = datetime(date_object.year, date_object.month, date_object.day)
    t1 = datetime.now()
    delta_t = t1 - t0
    return delta_t.total_seconds()

segundos_vividos('23/01/2021')
```
