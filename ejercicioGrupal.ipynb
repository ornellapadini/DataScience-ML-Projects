
# Registro de actividad física

Queremos desarrollar un programa en Python que nos permita registrar nuestras actividades físicas diarias y visualizar un resumen de los tiempos dedicados a cada tipo de actividad.

El programa debe cumplir con los siguientes requisitos:





1.   Permitir al usuario ingresar las actividades físicas realizadas y el tiempo dedicado a cada una en minutos.
2.   Almacenar los datos ingresados en una estructura de datos adecuada.
3.   Calcular y mostrar un resumen que incluya el tiempo total dedicado a todas las actividades y el porcentaje de tiempo dedicado a cada una.
4.   Mostrar un gráfico de barras que represente visualmente el porcentaje de tiempo dedicado a cada actividad.




```python
# Crear un diccionario para almacenar los datos de las actividades
actividades = {}

# Solicitar al usuario ingresar las actividades y los tiempos
while True:
    actividad = input("Ingrese una actividad física (o 'salir' para terminar): ")
    if actividad == "salir":
        break
    tiempo = float(input("Ingrese el tiempo dedicado a la actividad (en minutos): "))

    # Verificar si la actividad ya está en el diccionario
    if actividad in actividades:
        actividades[actividad] += tiempo
    else:
        actividades[actividad] = tiempo

# Calcular el tiempo total y el porcentaje de cada actividad
tiempo_total = sum(actividades.values())
porcentajes = {actividad: (tiempo / tiempo_total) * 100 for actividad, tiempo in actividades.items()}

# Mostrar un resumen de las actividades y sus porcentajes
print("\nResumen de actividades:")
for actividad, tiempo in actividades.items():
    porcentaje = porcentajes[actividad]
    print(f"{actividad}: {tiempo} minutos ({porcentaje:.2f}%)")

# Mostrar un gráfico de barras
import matplotlib.pyplot as plt

actividades_labels = list(actividades.keys())
tiempos = list(actividades.values())

plt.bar(actividades_labels, tiempos)
plt.xlabel('Actividades')
plt.ylabel('Tiempo (minutos)')
plt.title('Registro de actividades físicas')
plt.show()
```
